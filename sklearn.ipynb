```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
n =1000
cats = 6
dims = 2

x = []
y = []

for i in range(cats):
    x.append(np.random.randn(n, dims) + np.random.randn(dims)*10)
    y.append(np.array([np.arange(cats)] * n) ==i)

x = np.vstack(x)   
y = np.vstack(y) 

x_max = np.max(x, axis = 0)
x_min = np.min(x, axis = 0)   
x -=x_min
x /=x_max - x_min


plt.scatter(x[:,0], x[:,1], c=np.argmax(y, axis =1), cmap ='jet' )
    
```

```python
model = LogisticRegression()
model.fit(x, np.argmax(y, axis =1))
y_hat = model.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_hat, cmap = 'jet')
```
