```python

import torch
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Torch


```python
n = 1000
d = 2


x_0 = torch.randn(n, d) + torch.randn(1, d)*10
x_1 = torch.randn(n, d) + torch.randn(1, d)*-1

y_0 = torch.zeros((n,1))
y_1 = torch.ones((n,1))

x = torch.cat([x_0, x_1])
y = torch.cat([y_0, y_1])


plt.scatter(x[:,0], x[:,1],c=y.flatten(),  cmap ='jet' )

```

## Tensorflow

```python
n = 1000
d = 2

x_0 = tf.random.normal((n, d)) + tf.random.normal((1, d))*10
x_1 = tf.random.normal((n, d)) + tf.random.normal((1, d))*-1

y_0 = tf.zeros((n,1))
y_1 = tf.ones((n,1))

x = tf.concat([x_0, x_1], axis = 0)
y = tf.concat([y_0, y_1], axis = 0)



#plt.scatter(x[:,0], x[:,1],c=tf.squeeze(y),  cmap ='jet' )   #tf.squeeze(y) and tf.reshape(y, [-1]) is same
plt.scatter(x[:,0], x[:,1],c=tf.reshape(y,[-1]),  cmap ='jet' )
```
